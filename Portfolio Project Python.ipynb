```python
# Import Libraries

import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) # Adjusts configuration of the plots we will create

# Read in the data
df = pd.read_csv(r"C:\Users\nirav\Downloads\movies.csv")
```

```python
# Let's look at the data

df.head(10)
```

```python
# Let's see if there are any null values

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
    
    
df.isnull().sum()
```

```python
df.shape[0]
```

```python
df = df.dropna()
```

```python
df.shape[0]
```

```python
# Data Types for our Columns

df.dtypes
```

```python
# Change data types

df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
df['votes'] = df['votes'].astype('int64')
```

```python
df.sort_values(by = ['gross'], inplace = False, ascending = False)
```

```python
#pd.reset_option('display.max_rows')
```

```python
# Drop any duplicates

df['company'].drop_duplicates()
```

```python
# Hypotheses
# Budget has high correlation with gross
# Company will have high correlation
```

```python
# Scatterplot with budget vs gross

plt.scatter(x = df['budget'], y = df['gross'])
plt.title('Buidget vs Gross Earnings')
plt.xlabel('Budget')
plt.ylabel('Gross')

plt.show()
```

```python
# Regression plot 
sns.regplot(x = 'budget', y = 'gross', data = df, scatter_kws = {"color":"red"}, line_kws = {"color":"blue"})
```

```python
#Lets start looking at correlation
df.corr(method = 'pearson') # pearson, kenall, spearman
```
